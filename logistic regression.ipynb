```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
iris =  datasets.load_iris()
print(list(iris.keys()))
```

```python
print(iris['data'])
print(iris['target'])
```

```python
iris.DESCR.splitlines()
```

```python
# training data for logistic regression to predict weather a flower is iris virginica or not
import numpy as np

x = iris['data'][:,3:]              # 3rd column return garxa i.e petal width
y = (iris['target']==2).astype(np.int)
print(x)
y
```

```python
# train a logistic regression classifier

clf = LogisticRegression()
clf.fit(x,y)
example = clf.predict(([[2.6]]))              # predict ma hamro aba tyo petal width halne to predict 
example
```

```python
# plotting the graph
x_new = np.linspace(0,3,1000).reshape(-1,1)
y_prob = clf.predict_proba(x_new)
plt.plot(x_new,y_prob[:,1],'g-', label = "virginica")
```
